### Zadanie 4

Popraw kod ISO Włoch (ITA) wykorzystując komendy poznane na zajęciach o typie String.

Użyk kombinacji koment sortowania i odwracania listy, tak aby Polska (POL) znajdowała się na pierwszym miejscu. 

```python
lista_skroty = ["KOR", "POL", "GER", "FRA", "ita"]

#Przed poprawnym posortowaniem listy powstrzymuje nas zapis kodu WŁoch (ITA) małymi literami. 
#Zmieniamy to na wielkie - korzystamy z metody string upper:
lista_skroty[4] = lista_skroty[4].upper()
print("Kod Włoch po poprawieniu:")
print(lista_skroty[4], "\n")

#Nastepnie sortujemy liste - kod POL jest na koncu
lista_skroty.sort()

print("Lista po podstawowym sortowaniu:")
print(lista_skroty)

#Teraz wystarczy obrócić kolejność
lista_skroty.reverse()
print("\n","Końcowy wynik:", sep="")
print(lista_skroty)
```

### Zadanie 5

Poniższy kod wygeneruje listę losowych liczb. Długość listy będzie uzależniona od losowania z 5 długości. Będą się one wahać od 20 do 30 elementów:

```python
# Ten Kod nie wymaga waszej ingerencji - po prostu go uruchomcie
import random
rozmiary_list = [20, 23, 25, 27, 30]

dlugosc_listy = random.choice(rozmiary_list) 

losowa_lista = []
for i in range(0,dlugosc_listy):
  n = random.randint(1,30)
  losowa_lista.append(n)

print(losowa_lista)
```

Napisz taki kod, aby zawsze poprawnie drukował: 

* 3 od końca element listy 
* 5 i 6 od końca element listy

***Podpowiedzi:*** Użyj funkcji len() i ujemnego indeksowania 

```python
# Do rozwiązania można podejść na kilka sposobów - zacznihjmy od funkcji len()
print(losowa_lista[len(losowa_lista) - 3])

# W przypadku podawania zakresów trzeba podać indeks o jeden większy od tego który chcemy wziąć:
# Składnia wygląda następująco:
print(losowa_lista[len(losowa_lista) - 6: len(losowa_lista) - 4])

# Prościej jednak korzystać z ujemnego numerowania:
print(losowa_lista[-3])
print(losowa_lista[-6:-4])

# Oczywiście zachęcam do prostszych metod.
```
