Taka pętla to trzy linikki kodu - operujemy dwiema zmiennymi na kolekcji items.

### Zadanie 3 
Napisz program który policzy częstotliwości każdego wyrazu. Zwróć uwagę na fakt, że podczas przeglądania listy trzeba uwzględnić dwa warunki:
1. Elementu nie ma w słowniku i trzeba go dodać.
2. Element wystąpił wcześniej i trzeba odpowiadającą mu wartość słownika podnieść o 1. 

**Podpowiedź**: Istnienie klucza można sprawdzić podobnie jak w przypadku zbiorów. Taki kod:


```
# if "klucz" in dictionary
```

zwróci pożadany wynik.



````python
lista_wyrazow = "ala ma kota. kota ma ala. ala ma psa".split()
print("Wydrukujmy listę wyrazów:")
print(lista_wyrazow)
````

**Rozwiązanie:**

Zacznijmy od czegoś prostego najpierw zróbmy kod, który doda każdy nowy wyraz do złownika.

Wygląda on następująco: 

````python
# Tworzy pusty slownik
slownik_wyrazow = {}

# Pętla iteruje po wszystkich wyrazach z listy
for element in lista_wyrazow:
    slownik_wyrazow[element] = 1

print("")
print("Słownik unikatowych wyrazów prezentuje się następująco:")
print(slownik_wyrazow)
````

Teraz potrzebujemy dodać warunki, które psrawdzają czy dane słowo występowało już w słowniku. Jeżeli tak, powiększa licznik.

Rozbudowujemy kod z poprzedniego listingu:

````python
slownik_wyrazow = {}

for element in lista_wyrazow:

  # Każdy wyraz jest sprawdzany pod kątem wystapienia w słowniku
  if element in slownik_wyrazow:
     # Jeżeli element był - liczba jego wystąpień jest powiększana o 1 
     slownik_wyrazow[element] += 1
  else:
    # Jeżeli nie - ta linijka inicjuje nowy element z wartością 1
    slownik_wyrazow[element] = 1

print("")
print("Finalny słownik prezentuje się następująco:")
print(slownik_wyrazow)
````

### Zadanie 5
Wykorzystaj metode *type*, aby sprawdzić 

* Typ kolekcji *dict.items()*, *dict.keys()* oraz *dict.values()* dla słownika poniżej. 
* Typy elementów ze słownika w zależnosci od tego jaką kolekcje iterujemy w pętli.


````python
a = {"a": 1, "b":2}

# Tutaj sprawdzamy klasy z pierwszego punktu
print(type(a.items()), type(a.keys()), type(a.values()))

# Tutaj iterujemy w pętli po każdej z kolekcji z poprzedniego punktu:
print()
print("Typy - kolekcja elementy:")
for element in a.items():
  # Tu warto zwrócić uwagę, że nasze elementy to krotki zawierające pary: klucz - pozycja 0, wartość - pozycja 1
  print(type(element), element)

print()
print("Typy - kolekcja kluczy:")
for element in a.keys():
  print(type(element), element)

print()
print("Typy - kolekcja wartosci:")
for element in a.values():
  print(type(element), element)
````
