Napisz funkcję, która: 
* jako argument przyjmuje listę. Lista będzie zawsze zawierać ciąg cyfr zapisany w formacie string
* utworzy nową listę, gdzie każdy element będzie sumą cyfr wystepujących w ciągu znaków 
* zwróci tą listę 

Dla przykładu zaaplikowanie takiej funkcji do lista1 powinno dać nastepujący rezultat:
```
# [6, 18, 4]
```
Wydrukuj przerobione listy.


Podpowiedzi: 
* Konieczne będzie wykorzystanie dwóch pętli for w funkcji
* Przeszukiwanie łańcuchów znaków odbywać się może jak w liście albo po indeksach numerycznych - w tym drugim przypadku warto pamiętać o funkcji *len* zwracającej długość 


````python
# Gotowe rozwiązanie: 
lista1 = ["0123", "2556", "22"]
lista2 = ["0124", "2356", "122"]
lista3 = ["0125", "2456", "221"]
lista4 = ["0126", "2856", "222"]

def ZliczSumeCyfr(tekst):
    suma = 0
    #  pętla przegląda każdy znak w tekście
    for znak in tekst:
      # Ta linijka sumuje kolejne wartości 
      suma += int(znak)
    return suma 

def SumaLiczbWLiscie(listaWejsciowa):
  listaWynikowa = []
  
  # Pętla przegląda wszystkie elementy listy - każdy element to ciąg znaków
  for element in listaWejsciowa:
    suma = ZliczSumeCyfr(element)
    listaWynikowa.append(suma)
  return listaWynikowa

#Drukujemy wywołania dla kolejnych liczb
print(SumaLiczbWLiscie(lista1))
print(SumaLiczbWLiscie(lista2))
print(SumaLiczbWLiscie(lista3))
print(SumaLiczbWLiscie(lista4))
````

### Zadanie 2
Napisz funkcję, która:

* Jako argument przyjmuje listę pojedynczą liczbę
* Jeżeli liczba jest podzielna przez 3 zwraca jej wartość podniesioną do sześcianu
* W przeciwnym wypadku zwraca wartość podniesioną do kwadratu. 

Następnie wykorzystaj komendę *map*, aby stworzyć nową listę która zwróci efekty wywołania napisanej przez nas funkcji na przedstawionej poniżej liście liczb.

Podpowiedzi: 
* Do sprawdzenia podzielności wykorzystaj operator modulo (%)
* Składnia tworzenia nowej listy za pomocą funkcji *map* poniżej:

```
# NowaLista = list(map(Funkcja, Początkowa_lista))
```





````python
listaLiczb = [3, 5 , 9, 4]

# Nasza funkcja wygląda następującą:
def podniesieniePotega(liczba):
  if liczba%3 == 0:
    return liczba**3
  else:
    return liczba**2

# W tym momencie aplikuje ją do wszystkich elememntów listy za pomocą funkcji map:
NowaLista = list(map(podniesieniePotega, listaLiczb))
print(NowaLista)
````

### Zadanie 3 
Przygotuj wyrażenie lambda, które:
* jeżeli argument jest ciągiem znaku zwróci jego długość.
* w przeciwnym wypadku zwróci tą wartość.

Podpowiedzi:
* Wykorzystaj funkcję *isinstance*
* Składnia lambdy z wyrażeniem logicznym wygląda następująco:
```
# lambda argument: Wartosc_Prawda if Warunek else Wartosc_Falsz
```

Wydrukuj wynik zastosowania wyrażenia dla zmiennych zdefiniowanych poniżej:




````python
tekst = "Przykładowy tekst"
liczba = 2 

# Naszym warunkiem jest zapytanie czy typ zmiennej a to string - do tego służy funkcja isinstance
f = lambda a: len(a) if isinstance(a,str) else a

print("W przypadku tekstu drukowany jest rozmiar: ", f(tekst))
print("W przypadku liczby drukowana jest jej wartość: ",f(liczba))
````

### Zadanie 4
Napiszemy funkcję rekurencyjną, która wyznacza Największy Wspólny Dzielnik (NWD) dwóch liczb za pomocą algorytmu Euklidesa.

Algorytm można streścić następująco:
* Argumentem są dwie liczby - oznaczmy przez a i b
* Jeżeli a > b wykonaj funkcję NWD z argumentami (a-b, b)
* Jeżeli a < b wykonaj funkcję NWD z argumentami (a, b - a)
* Jeżeli a == b zwróć wartość a



````python
Liczba1 = 63
Liczba2 = 72

def NWD(a,b):
  if a > b: 
    return NWD(a-b,b)
  if b > a: 
    return NWD(a,b-a)
  else:
    return a

print(NWD(Liczba1, Liczba2))    
````
